## Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data

```python
time.sleep(2)
driver.get('https://www.naukri.com/data-scientist-jobs-in-bangalore?k=data%20scientist&l=bangalore')
```

```python
Job_Title= []
for d in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    Job_Title.append(d.text)
Job_Title[:10]   
```

```python
Job_Location=[]
for d in driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span"):
    Job_Location.append(d.text)
Job_Location[:10]
```

```python
Company_Name=[]
for d in driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']"):
    Company_Name.append(d.text)
Company_Name[:10]   
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute('href'))
urls
```

```python
Job_Description=[]
for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    # fetching job description
    
    try:
        job_desc=driver.find_element_by_xpath("//section[@class='job-desc']")
        Job_Description.append(job_desc.text.replace('\n'," "))
        
    except NoSuchElementException:
        Job_Description.append('_')
```

```python
Job_Description
```

```python
info_=list(zip(Job_Title,Job_Location,Company_Name,Job_Description))
Job_info_=pd.DataFrame(info_[:10],columns=['Job_Title','Job_Location','Company_Name','Job_Description'])
```

```python
Job_info_
```
